# Pipeline:

It is mechanism in which it chains multiple steps together and input of one step is output of next step.

These makes it easy to use same preprocessing steps on training and testing.

```python
import numpy as np
import pandas as pd
```

```python

df=pd.read_csv("C:\\Users\\user\\Datasets\\covid_toy.csv")
```

```python
df.head()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
```

```python
x=df.drop(columns=['has_covid'],axis=1)
y=df['has_covid']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
c_feature=['gender','cough','city']
n_feature=['age','fever']
```

```python
trf1=Pipeline(steps=[
    ('imputer',SimpleImputer(strategy='mean')),
    ('scaler',StandardScaler())
])
```

```python
trf2=Pipeline(steps=[
    ('imputer',SimpleImputer(strategy='most_frequent')),
    ('onehot',OneHotEncoder(handle_unknown='ignore'))
])
```

```python
preprocessor=ColumnTransformer(transformers=[
    ('num',trf1,n_feature),
    ('cat',trf2,c_feature)
])
```

```python

clf=Pipeline(steps=[
    ('preprocessor',preprocessor),
    ('classification',LogisticRegression())
])
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_pred,y_test))
```

# Principle Component Analysis:

It is unsupervised ml algorithm used to reduce higher dimension data to lower without changing its performance.

```python
df=pd.read_csv("C:\\Users\\user\\Datasets\\Social_Network_Ads.csv")
```

```python
df.head()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
df=df.drop(columns=['User ID','Gender'])
```

```python
x=df.drop(columns=['Purchased'],axis=1)
y=df['Purchased']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
pipe=Pipeline(steps=[
    ('scaler',StandardScaler()),
    ('pca',PCA(n_components=2)),
    ('classifier',RandomForestClassifier(n_estimators=100,random_state=42))
])
```

```python
pipe
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred))
```
